```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import make_regression
import plotly.express as px
import plotly.graph_objects as go
```

```python
X,y = make_regression(n_samples=300, n_features=2,n_informative=2,n_targets=1,noise=50)
```

```python
df = pd.DataFrame({'feature 1':X[:,0],'feature 2':X[:,1],'target':y})
```

```python
fig = px.scatter_3d(df, x='feature 1',y='feature 2',z='target')
fig.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSA",mean_squared_error(y_test,y_pred))
print("R2S",r2_score(y_test,y_pred))
```
